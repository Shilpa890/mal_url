```python
import pandas as pd
import numpy as np
url_data=pd.read_csv("C:/Work/Ques/url_interview_cleaned.csv",sep='\t', names=['label', 'url'])
url_data
```

```python
x=url_data.url
y=url_data.label
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vector=CountVectorizer(strip_accents='unicode',stop_words='english',ngram_range=(1,1),max_df=0.95,min_df=0.01)
```

```python
X_total=vector.fit_transform(x.values.astype('U'))
```

```python
X_df=pd.DataFrame(X_total.toarray(),columns=vector.get_feature_names())
```

```python
X_df
```

```python
X_df.dtypes
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_df,y,random_state=1)
```

```python
X_train
```

```python
X_train.dtypes
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
nb=MultinomialNB()
nb.fit(X_train,Y_train)
```

```python
y_train_pred=nb.predict(X_train)
```

```python
y_train_pred
```

```python
Y_train
```

```python
y_test_pred=nb.predict(X_test)
```

```python
y_test_pred=y_test_pred.astype(object)
```

```python
Y_test
```

```python
from sklearn import metrics
```

```python
metrics.roc_auc_score(Y_test,y_test_pred)
```

```python
metrics.accuracy_score(Y_test,y_test_pred)
```
